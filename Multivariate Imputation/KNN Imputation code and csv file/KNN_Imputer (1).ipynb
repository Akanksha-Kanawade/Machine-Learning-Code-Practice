 **We are working on KNN Imputer which is use to fill numeric value hence , we will consider only Numeric column.**


```python
data=pd.read_csv("/content/train.csv",usecols=['Age','Pclass','Fare','Survived'])
```

# Split the data

```python
x=data.drop(columns='Survived')
y=data['Survived']
```

```python
x.head(2)
```

```python
y.head(2)
```

# Do the Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python

from sklearn.impute import KNNImputer,SimpleImputer
```

```python
'''
KNN= KNNImputer(n_neighbors=2) # By changing neighbours we can change the accuracy
x_tain_trf=KNN.fit_transform(x_train)
x_test_trf=KNN.transform(x_test)
'''
```

```python
KNN= KNNImputer(n_neighbors=2,weights='distance') # By default weights=uniform  by changing weights it may increase accuracy
x_tain_trf=KNN.fit_transform(x_train)
x_test_trf=KNN.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()

model.fit(x_tain_trf,y_train)

y_pred=model.predict(x_test_trf)

from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```

**We multiply the values with its 1/distance to fill missing**

**Benifit of this is, it will give more weightage to distance near to that point**
